## Vortex beam passing through an circular aperture

```python
x0 = np.linspace(-200*um, 200*um, 1024)
y0 = np.linspace(-200*um, 200*um, 1024)
wavelength = 0.6238 * um



# Generaci√≥n de la clase
u0 = Scalar_source_XY(x=x0, y=y0, wavelength=wavelength)
# carga de la onda esferica
u0.vortex_beam(r0=(0,0), w0=60*um, m=4)

t0=Scalar_mask_XY(x=x0, y=y0, wavelength=wavelength)
t0.circle(r0=(0,0), radius=(150*um,150*um))

u1=u0*t0

u2=u1.RS(z=15*mm, new_field=True, verbose=True);
u2.normalize()
```
