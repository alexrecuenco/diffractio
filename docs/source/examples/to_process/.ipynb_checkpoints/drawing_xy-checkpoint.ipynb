## Vortex beam passing through an circular aperture

```python
x0 = np.linspace(-125*um, 125*um, 1024)
y0 = np.linspace(-125*um, 125*um, 1024)
wavelength = 0.6238 * um



# Generaci√≥n de la clase
u0 = Scalar_source_XY(x=x0, y=y0, wavelength=wavelength)
# carga de la onda esferica
u0.vortex_beam(r0=(0,0), w0=50*um, m=3)

t0 = Scalar_mask_XY(x=x0, y=y0, wavelength=wavelength)
t0.circle(r0=(0,0), radius=(80*um, 80*um))
u1=u0*t0

u1.RS(z=.5*mm, new_field=False, verbose=True);
u1.normalize()
```

```python
u0.draw()
```
