```python
from diffractio import degrees, mm, plt, sp, um, np
from diffractio.scalar_fields_XY import Scalar_field_XY
from diffractio.utils_drawing import draw_several_fields
from diffractio.scalar_masks_XY import Scalar_mask_XY
from diffractio.scalar_sources_XY import Scalar_source_XY
```

```python
def difraccion_biprism_Fresnel(z=1*mm):
    length = 200 * um
    numdatos = 1024
    x0 = np.linspace(-length / 2, length / 2, numdatos)
    y0 = np.linspace(-length / 2, length / 2, numdatos)
    wavelength = 0.6238 * um

    u1 = Scalar_source_XY(x=x0, y=y0, wavelength=wavelength)
    u1.plane_wave()

    t1 = Scalar_mask_XY(x=x0, y=y0, wavelength=wavelength)
    t1.circle(r0=(0 * um, 0 * um), radius=(100 * um, 100 * um))

    t2 = Scalar_mask_XY(x=x0, y=y0, wavelength=wavelength)
    t2.biprism_fresnel(
        r0=(0 * um, 0 * um), ancho=100 * um, height=5 * um, n=1.5)
    
    u3 = t1 * t2 * u1
    
    u4 = u3.RS(z=z, new_field=True)
    u4.draw(kind='intensity')
    plt.title("$z={:2.2f}\,mm$".format(z), fontsize=24)
```

```python
difraccion_biprism_Fresnel(z=0.05*mm)
```
