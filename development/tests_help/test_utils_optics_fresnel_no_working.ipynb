# Fresnel equations

```python
from diffractio import degrees, mm, nm, np, plt, sp, um
```

```python
from diffractio.utils_optics import fresnel_coefficients_dielectric, reflectance_transmitance_dielectric
from diffractio.utils_optics import draw_fresnel_coefficients, draw_transmittance, reflectance_transmitance_complex
```

##    interfaz aire-vidrio

```python
n1 = 1.4
n2 = 1.5
theta_i = np.linspace(0 * degrees, 90 * degrees, 100)

# se calculan los coeficientes de Fresnel para un array
r_perp, r_par, t_perp, t_par = fresnel_coefficients_dielectric( theta_i=theta_i,
    n1=n1, n2=n2)


R_perp, R_par, T_perp, T_par = reflectance_transmitance_dielectric(theta_i, 
    n1, n2)
```

```python
draw_fresnel_coefficients(
    theta_i, n1, n2, r_perp, r_par, t_perp, t_par, filename='')
```

```python
draw_transmittance(theta_i, n1,  n2, R_perp, R_par, T_perp, T_par)
```

```python
n1 = 1.
n2 = 1. - 0.5j

theta_i = np.linspace(0 * degrees, 90 * degrees, 100)

r_perp, r_par, t_perp, t_par = reflectance_transmitance_complex(theta_i=theta_i,
    n1=n1,  n2c=n2)

R_perp, R_par, T_perp, T_par = reflectance_transmitance_dielectric(theta_i,
    n1,  n2)
```
