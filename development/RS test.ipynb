## XY

```python
from diffractio import np, plt, um, mm, nm, degrees
from diffractio.scalar_masks_XY import Scalar_mask_XY
from diffractio.scalar_sources_XY import Scalar_source_XY
```

```python
from matplotlib import rcParams
rcParams['figure.dpi']=200
rcParams['font.size']=14
```

```python
x0=np.linspace(-200*um, 200*um, 256)
y0=np.linspace(-200*um, 200*um, 256)
wavelength=532*nm
```

```python
u0=Scalar_source_XY(x0,y0,wavelength)
u0.gauss_beam(r0=(0,0), w0=130*um, z0=0, A=1, theta=0.0)
```

```python
obstacle=Scalar_mask_XY(x0,y0,wavelength)
obstacle.circle(r0=(0,0), radius=50*um)
#obstacle.ring(r0=(0,0), radius1=(50,50), radius2=(115,115))
obstacle.u=1-obstacle.u

u1=u0*obstacle
```

```python
arago_point=u1.RS(z=2*mm, verbose=True)
```

```python
arago_point.draw(logarithm=True);
```

```python
arago_point2=u1.RS(z=20*mm, amplification=(2,2), verbose=True, matrix=True)
plt.imshow(np.abs(arago_point2))
plt.set_cmap('jet')
```

```python
arago_point2=u1.RS(z=25*mm, amplification=(2,2), verbose=True)
arago_point2.draw();
```

```python
arago_point2=u1.RS(z=20*mm, amplification=(4,2), verbose=True)
arago_point2.draw();
```
